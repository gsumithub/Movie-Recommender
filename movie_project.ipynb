```python
movies=movies[['id', 'title', 'genre', 'overview']]
```

```python
movies
```

```python
movies.loc[:, 'tags'] = movies['overview'].astype(str) + ' ' + movies['genre'].astype(str)
```

```python
movies
```

```python
new_data = movies.drop(columns=['overview', 'genre'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title'] == "The Godfather"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    # Find the index of the movie in the DataFrame
    index = new_data[new_data['title'] == movies].index[0]
    
    # Calculate the cosine similarity distance
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])
    
    # Print the titles of the top 5 similar movies
    for i in distance[1:6]:  # Starting from 1 to avoid recommending the same movie
        print(new_data.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```

```python
recommend("Batman")
```

```python
recommend("The Godfather")
```
